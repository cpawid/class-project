# CARAT,CUT,CLARITY, COLOR RELATIONSHIP WITH PRICE

```python
#does carat affect the price?
carat_price = diamond.groupby('carat')['price'].describe()
carat_price
```

```python
carat_price = diamond.groupby('carat')['price'].aggregate(np.mean).sort_values(ascending=False)
carat_price
```

```python
depth_price = diamond.groupby('depth')['price'].describe()
depth_price
```

```python
color_price = diamond.groupby('color')['price'].mean()
color_price
```

```python
clarity_price = diamond.groupby('clarity')['price'].mean()
clarity_price
```

```python
#analysis for carat and price
sns.scatterplot(x=diamond.carat, y=diamond.price)
plt.title('Price Correlation in Carat',
          fontsize = 15, color = 'Black')
plt.show()
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(diamond['clarity'],diamond['price'],ax=axis1).set_title('Price Correlation on Clarity')
sns.barplot(diamond['color'],diamond['price'],ax=axis2).set_title('Price Correlation on Color')
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(diamond['carat'],diamond['price'],ax=axis1).set_title('Price Correlation on Carat')
sns.barplot(diamond['cut'],diamond['price'],ax=axis2).set_title('Price Correlation on Cut')
```
